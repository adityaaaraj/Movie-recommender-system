```python

def extract_values_from_column(df, column_name, key):
    
    def extract_values(json_str):
        try:
            # Ensure the JSON-like string is properly formatted
            json_str = '[' + json_str.strip().strip('][') + ']'
            # Parse the JSON-like string
            data_list = json.loads(json_str)
            # Extract the values associated with the specified key
            values = [item[key] for item in data_list]
            return values
        except json.JSONDecodeError:
            print("Invalid JSON string")
            return []
        
    return df[column_name].apply(extract_values)
    
   
```

```python
movies['genres'] = movies['genres'].astype(str)
movies['genres'] = extract_values_from_column(movies, 'genres', 'name')
```

```python
import ast
```

```python
def convert(obj):
    L = []
    cnt = 0
    for i in ast.literal_eval(obj):
        if cnt != 3:
            L.append(i['name'])
            cnt += 1
        else:
            break
    return L
             
```

```python
movies['cast_name'] = movies['cast'].apply(convert)
```

```python
def get_first_three_names(name_list):
    
    return name_list[:3]
```

```python
movies['cast'] = movies['cast_name'].apply(get_first_three_names)
```

```python
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]

```

```python
movies.head()
```

```python
def fetch_director(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
             
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))

    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features = 5000, stop_words = 'english')
```

```python
tag_matrix = cv.fit_transform(new_df['tags'])
```

```python
vectors = tag_matrix.toarray()
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity #0 se 1 ke beech number aega,inv of cosine distance i.e. angle between vectors
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
def recommend(movie):
    movie_index = new_df[new_df['title'] == movie].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse = True, key=lambda x:x[1])[1:6]

    for i in movies_list:
        print(new_df.iloc[i[0]].title)
```

```python
recommend('Avatar')
```
